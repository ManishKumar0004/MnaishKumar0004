### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results ?
### Ans :-

**In this question we are given  ,<br>**

- **The mean sample is , , x̅ = 50<br>**
- **The sample standard deviation is,  s = 5<br>**
- **We are assuming , Sample size, n = 50<br>**
- **Confidence level is 95% or 0.95:<br>**
- **The remaining probability (α) is then: 5%, or 1 - 0.95 = 0.05.<br>**

**We need to estimate the population mean revenue with a 95% confidence interval.<br>**

### Calculating a Confidence Interval
**The following steps are used to calculate a confidence interval:<br>**

**(i) Check the conditions<br>**
**(ii) Find the point estimate<br>**
**(iii) Decide the confidence level<br>**
**(iv) Calculate the margin of error<br>**
**(v) Calculate the confidence interval<br>**


### 1. Checking the Conditions
**The conditions for calculating a confidence interval for a mean are:<br>**

**The sample is randomly selected<br>**
**And either:<br>**
**The population data is normally distributed<br>**
**Sample size is large enough<br>**
**A moderately large sample size, like 50, is typically large enough.<br>**

**In the example, the sample size was 50 and it was randomly selected, so the conditions are fulfilled.<br>**

### 2. Finding the Point Estimate
**The point estimate is the sample mean.<br>**
**x̅ = 50**

### 3. Deciding the Confidence Level
**The confidence level is expressed with a percentage or a decimal number.<br>**
**the confidence level is 95% or 0.95:<br>**
**The remaining probability (α) is then: 5%, or 1 - 0.95 = 0.05.<br>**

### 4. Calculating the Margin of Error
**The margin of error is the difference between the point estimate and the lower and upper bounds.<br>**
**The margin of error (E) for a proportion is calculated with a critical t-value and the standard error:**

**In this question with a sample standard deviation (s
) of 5 and sample size of 50 the standard error is**:

**s / √n = 5 / √50  = 0.7071**

**So we need to find the critical t-value**

**The critical t-value can be found using a t-table or with a programming language function:**

**E  = 2.010 * 0.7071**

**E = 1.421**

### 5. Calculate the Confidence Interval
**The lower and upper bounds of the confidence interval are found by subtracting and adding <br>the margin of error E
) from the point estima (x_bar).**

**The lower bound is:**

#### x̅ - E = 50 - 1.421 = 48.579

**The upper bound is:**

#### x̅ - E = 50 + 1.421 = 51.421

**The confidence interval is:**

### C.I = [48.579 , 51.421]

**The 95% confidence interval for the sample mean is approximately (48.579,51.421). This means we are 95% confident that the true population mean lies within this interval.**

###  Python function to estimate the population mean 

```python
import scipy.stats as stats
import math

# Specify sample mean (x_bar), sample standard deviation (s), sample size (n) and confidence level
x_bar = 50
s = 5
n = 50
confidence_level = 0.95

# Calculate alpha, degrees of freedom (df), the critical t-value, and the margin of error
alpha = (1-confidence_level)
df = n - 1
standard_error = s/math.sqrt(n)
critical_t = stats.t.ppf(1-alpha/2, df)
margin_of_error = critical_t * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Print the results
print("Critical t-value: {:.3f}".format(critical_t))
print("Margin of Error: {:.3f}".format(margin_of_error))
print("Confidence Interval: [{:.3f},{:.3f}]".format(lower_bound,upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.3f} and {:.3f}".format(lower_bound,upper_bound))
```

**The Chi-Square Goodness-of-Fit test is a non-parametric test. Non-parametric tests<br> do not assume that the data comes from a specific probability distribution or make any assumptions <br>about population parameters like the mean or standard deviation.**

### Assumptions
**(1) Independence: The observations in the sample must be independent
of each other. This means that the outcome of one observation
should not influence the outcome of another observation.<br>**

**(2) Categorical data: The variable being analysed must be categorical,

exclusive and exhaustive categories.<br>**

**(3) Expected frequency: Each category should have an expected
frequency of at least 5. This guideline helps ensure that the Chi-
Square distribution is a reasonable approximation for the
distribution of the test statistic. Having small expected frequencies
can lead to an inaccurate estimation of the Chi-Square distribution,
potentially increasing the likelihood of a Type I error (incorrectly
rejecting the null hypothesis) or a Type II error (incorrectly failing to
reject the null hypothesis).<br>**

**(4) Fixed distribution: The theoretical distribution being compared to
the observed data should be specified before the test is conducted.
It is essential to avoid choosing a distribution based on the observed
data, as doing so can lead to biased results.<br>**

###  Python function to estimate the true population proportion

```python
import scipy.stats as stats
import math

# Given data
n = 500  
x = 60   
# Sample proportion
p_hat = x / n

# Calculate standard error (SE)
SE = math.sqrt((p_hat * (1 - p_hat)) / n)

# Calculate the margin of error (ME) using the Z-score for 95% confidence
Z = stats.norm.ppf(0.975)  
ME = Z * SE

# Calculate the confidence interval
lower_bound = p_hat - ME
upper_bound = p_hat + ME

# Print the results
print(f"Sample proportion (p_hat): {p_hat}")
print(f"95% Confidence Interval for the true proportion of smokers: ({lower_bound}, {upper_bound})")
```

### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15 ?
### Ans :- 

### Plot a Single Chi-Square Distribution with 10 degrees of freedom

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Generate x values for the chi-squared distribution
x = np.linspace(0, 30, 1000)

# Degrees of freedom for the chi-squared distribution
df = 10

# Compute the chi-squared PDF values
y = chi2.pdf(x, df)

plt.figure(figsize=(6,4))

# Plot the chi-squared distribution
plt.plot(x, y, label=f'Chi-squared Distribution (df={df})')

# Shade the area under the curve where x is between 0 and 15
plt.fill_between(x, y,
                 where=(x > 0) & (x <= 15),
                 color='skyblue', alpha=0.5, label='Shaded Region')

# Add labels and title
plt.xlabel('Chi-squared statistic')
plt.ylabel('Probability density')
plt.title('Chi-squared Distribution')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
