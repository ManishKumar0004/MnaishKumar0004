### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test ?
### Ans :-

### Z-Test :
**The Z-test is a statistical test used to determine if there is a significant difference between a sample mean and a population mean. It is commonly employed when dealing with large sample sizes (typically equal to or greater than 30) and situations where the population standard deviation is known. By comparing the Z-statistic to the critical value, the Z-test helps assess the likelihood of the observed difference between the means being statistically significant.**

### Advantages of Z-Test
- **`Suitable for large sample sizes`: The Z-test is specifically designed for situations where the sample size is large, making it an appropriate choice for analyzing data with a sufficient number of observations.<br>**

- **`Simplicity`: The Z-test is relatively simpler to implement compared to other statistical tests, requiring fewer calculations and assumptions.<br>**

- **`Known population standard deviation`: The Z-test assumes a known population standard deviation, which can be advantageous when working with data that meets this assumption.<br>**

- **`Provides precise p-values`: The Z-test provides more precise p-values compared to other tests, allowing for more accurate assessments of statistical significance.<br>**

- **`Standard normal distribution`: The Z-test utilizes the standard normal distribution (Z-distribution), which is well-defined and widely understood in statistical theory.<br>**

- **`Robustness`: The Z-test is robust to violations of the normality assumption, particularly when the sample size is large enough to satisfy the central limit theorem.<br>**

- **`Allows for hypothesis testing`: The Z-test enables researchers to test specific hypotheses about population means, providing statistical evidence to support or reject these hypotheses.<br>**

- **`Reliable for large effect sizes`: The Z-test is reliable for detecting large effect sizes, making it suitable for scenarios where substantial differences are expected.<br>**

- **`Widely recognized and accepted`: The Z-test is a well-established statistical test with a long history of usage, making it widely recognized and accepted in the scientific community.<br>**

- **`Enables comparisons with known population parameters`: The Z-test allows for comparisons between sample means and known population parameters, facilitating more informed decision-making.<br>**

### T-Test :
**The T-test is a statistical test used to determine if there is a significant difference between the means of two groups or samples. It is commonly employed when dealing with small sample sizes (usually less than 30) and unknown population standard deviations. By calculating the T-statistic and comparing it to the critical value, the T-test helps assess the probability that the observed difference between the means is due to chance or a genuine effect.**

### Advantages of T-Test
- **`Suitable for small sample sizes`: The T-test is specifically designed for situations where the sample size is small, making it an ideal choice for analyzing data in such cases.<br>**
- **`Flexible application`: The T-test can be applied to various scenarios, including one-sample comparisons, independent two-sample comparisons, and paired two-sample comparisons.<br>**
- **`Accounts for variation`: The T-test takes into account the variability within each group or sample, allowing for more accurate comparisons.<br>**
- **`Robustness`: The T-test is relatively robust to violations of the normality assumption, making it a reliable statistical test even when the data is not perfectly normally distributed.<br>**
- **`Handles unequal variances`: The T-test has modifications such as the Welch's T-test that can handle unequal variances between groups, providing accurate results in such cases.<br>**
- **`Widely recognized`: The T-test is a well-known and widely accepted statistical test, making it easier to interpret and communicate the results to others.<br>**
- **`Provides a measure of confidence`: The T-test provides a measure of confidence in the significance of the observed differences between group means, allowing researchers to draw meaningful conclusions.<br>**
- **`Offers a standardized effect size`: The T-test can be accompanied by effect size measures such as Cohen's d, which quantifies the magnitude of the observed differences.<br>**
- **`Enables hypothesis testing`: The T-test allows researchers to test specific hypotheses regarding mean differences between groups, providing statistical evidence to support or reject these hypotheses.<br>**
- **`Availability of software packages`: There are numerous software packages and statistical tools that offer built-in functions for conducting T-tests, simplifying the analysis process.<br>**

### Q2: Differentiate between one-tailed and two-tailed tests ?

### Ans :-

**(1) One Tailed Hypothesis Testing :<br>**

**One tailed hypothesis testing is done when the rejection region is only in one direction. It can also be known as directional hypothesis testing because the effects can be tested in one direction only.<br>**

**This type of testing is further classified :<br>**

- **Right Tailed Hypothesis Testing<br>**
- **Left Tailed Hypothesis Testing**

**(2) Two Tailed Hypothesis Testing :<br>**
**In this hypothesis testing method, the critical region lies on both sides of the sampling distribution. It is also known as a non - directional hypothesis testing method.**


**Whenever we perform a hypothesis test, we always write a null hypothesis and an alternative hypothesis, which take the following forms:<br>**

**H0 (Null Hypothesis): Population parameter =,  ≤, ≥ some value<br>**

**HA (Alternative Hypothesis): Population parameter <, >, ≠ some value<br>**

**`Note`:- Note that the null hypothesis always contains the equal sign.<br>**

### Example : Height of Males
**It’s assumed that the mean height of males in a certain city is 68 inches. However, an independent researcher believes the true mean height is greater than 68 inches. To test this, he goes out and collects the height of 50 males in the city.<br>**

**Here is how to write the null and alternative hypotheses for this scenario:<br>**

**H0: μ ≤ 68 (the true mean height is equal to or even less than 68 inches)<br>**

**H1: μ > 68 (the true mean height is greater than 68 inches)<br>**

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error?
### Ans :-


**In hypothesis testing, there are two types of errors that can occur <br>when making a decision about the null hypothesis: <br>**
- **Type I error<br>**
- **Type II error<br>**

### Type I error :
**Type-I (False Positive) error occurs when the sample results, lead to the rejection of the <br>null hypothesis when it is in fact true.<br>**

**In other words, it's the mistake of finding a significant effect or relationship<br> when there is none. The probability of committing a Type I error is denoted by α (alpha), which is also known as the significance level. By choosing a significance level, researchers can control the risk of making a Type I error.<br>**

**To reduce the Type I error probability, you can simply set a lower significance level.**

### Type I error rate
- **The null hypothesis distribution curve below shows the probabilities of obtaining all possible results if the study were repeated with new samples and the null hypothesis were true in the population.<br>**

- **At the tail end, the shaded area represents alpha. It’s also called a critical region in statistics.<br>**

- **If your results fall in the critical region of this curve, they are considered statistically significant and the null hypothesis is rejected. However, this is a false positive conclusion, because the null hypothesis is actually true in this case!<br>**

### Examples of Type I Error 
- **Legal System: In a criminal trial, the null hypothesis (H0) is that the defendant is innocent, while the alternative hypothesis (H1) is that the defendant is guilty. A Type I error occurs if the jury convicts the defendant (rejects the null hypothesis) when they are actually innocent.<br>**

- **Medical Testing: Suppose a medical test is designed to diagnose a particular disease. The null hypothesis (H0) is that the person does not have the disease, and the alternative hypothesis (H1) is that the person does have the disease. A Type I error occurs if the test incorrectly indicates that a person has the disease (rejects the null hypothesis) when they do not actually have it.<br>**

### Type II error :
**Type-II (False Negative) error occurs when based on the sample results,<br>the null hypothesis is not rejected when it is in fact false.<br>**

**This means that the researcher fails to detect a significant effect or relationship<br> when one actually exists. The probability of committing a Type II error is denoted by β (beta).<br>**

**To (indirectly) reduce the risk of a Type II error, you can increase the sample size or the significance level.**

### Type II error rate
**The null hypothesis distribution curve below shows the probabilities of obtaining all possible results if the study were repeated with new samples and the null hypothesis were true in the population.<br>**

**At the tail end, the shaded area represents alpha. It’s also called a critical region in statistics.<br>**

**If your results fall in the critical region of this curve, they are considered statistically significant and the null hypothesis is rejected. However, this is a false positive conclusion, because the null hypothesis is actually true in this case!<br>**

### Examples of Type II Error
- **Legal System: In a criminal trial, a Type II error occurs if the jury acquits the defendant (fails to reject the null hypothesis) when they are actually guilty.<br>**

- **Medical Testing: In a medical test designed to diagnose a disease, a Type II error occurs if the test incorrectly indicates that a person does not have the disease (fails to reject the null hypothesis) when they actually do have it.**


### Factors Affecting Type I and Type II Errors
#### Some of the common factors affecting errors are:

- **Sample Size: In statistical hypothesis testing, larger sample sizes generally reduce the probability of both Type I and Type II errors. With larger samples, the estimates tend to be more precise, resulting in more accurate conclusions.
<br>**

- **Significance Level: The significance level (α) in hypothesis testing determines the probability of committing a Type I error. Choosing a lower significance level reduces the risk of Type I error but increases the risk of Type II error, and vice versa.<br>**
  
- **Effect Size: The magnitude of the effect or difference being tested influences the probability of Type II error. Smaller effect sizes are more challenging to detect, increasing the likelihood of failing to reject the null hypothesis when it’s false.<br>**
  
- **Statistical Power: The power of Statistics (1 – β) dictates that the opportunity for rejecting a wrong null hypothesis is based on the inverse of the chance of committing a Type II error. The power level of the test rises, thus a chance of the Type II error dropping.<br>**

- **Box A contains 2 red and 3 black balls<br>**
- **Box B contains 3 red and 1 black ball<br>**
- **And box C contains 1 red ball and 4 black balls<br>**

**All the three boxes are identical having an equal probability to be picked up. Therefore, what is the probability that the red ball was picked up from box A?<br>**

### Solution

**Let E denote the event that a red ball is picked up and A, B and C denote that the ball is picked up from their respective boxes. Therefore the conditional probability would be P(A|E) which needs to be calculated.<br>**

**The existing probabilities P(A) = P(B) = P (C) = 1 / 3, since all boxes have equal probability of getting picked.<br>**

**P(E|A) = Number of red balls in box A / Total number of balls in box A = 2 / 5<br>**

**Similarly, P(E|B) = 3 / 4 and P(E|C) = 1 / 5<br>**

**Then evidence P(E) = P(E|A) * P(A) + P(E|B) * P(B) + P(E|C) * P(C)<br>**

**&emsp; &emsp; &emsp; &emsp; &emsp;&emsp; &emsp;  = (2/5) * (1/3) + (3/4) * (1/3) + (1/5) * (1/3) = 0.45<br>**

**Therefore, P(A|E) = P(E|A) * P(A) / P(E) = (2/5) * (1/3) / 0.45 = 0.296<br>**

### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example ?
### Ans :-

### Confidence interval :
**Definition : "A confidence interval is the range of values, derived from sample statistics, that is likely to contain the value of an unknown population parameter with a certain degree of confidence."<br>**

### Interpreting Confidence Interval :-
**A confidence interval is a range of values within which a population parameter, such as the
population mean, is estimated to lie with a certain level of confidence. The confidence interval
provides an indication of the precision and uncertainty associated with the estimate. To
interpret the confidence interval values, consider the following points:<br>**


**1. Confidence level: The confidence level (commonly set at 90%, 95%, or 99%) represents
the probability that the confidence interval will contain the true population parameter if
the sampling and estimation process were repeated multiple times. For example, a 95%
confidence interval means that if you were to draw 100 different samples from the
population and calculate the confidence interval for each, approximately 95 of those
intervals would contain the true population parameter.<br>**

**2. Interval range: The width of the confidence interval gives an indication of the precision of
the estimate. A narrower confidence interval suggests a more precise estimate of the
population parameter, while a wider interval indicates greater uncertainty. The width of
the interval depends on the sample size, variability in the data, and the desired level of
confidence.<br>**

**3. Interpretation: To interpret the confidence interval values, you can say that you are "X%
confident that the true population parameter lies within the range (lower limit, upper
limit)." Keep in mind that this statement is about the interval, not the specific point
estimate, and it refers to the confidence level you chose when constructing the interval.<br>**
