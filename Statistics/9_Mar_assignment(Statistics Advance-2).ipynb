### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example ?
### Ans :-

### Probability Mass Function (pmf) :
- **A probability mass function (pmf) is a mathematical function that calculate the probability of a discrete <br>random variable.<br>**
- **A probability mass function gives the probability that a discrete random variable is exactly equal to <br>some value.<br>**
- **A discrete random variable can take on a finite or countably infinite number of possible values.<br>**
- **The standard notation for a probability mass function is given below : <br>**

**here,<br>**
**--> X is the discrete random variable<br>**
**--> x is one of the possible discrete values<br>**
**--> f(x) is a mathematical function that calculate the probability for the value of x**

- **So , putting it all together , P(X = x) = f(x) means , the chance of variable X assuming the<br> specific value of x equals f(x)<br>**
- **In pmf, the probabilities for all possible values must sum to one.Because the chance is 1.**

### Example :-
**(1) Imagine flipping a fair coin<br>**
**(2) Rolling a dice , the chance of rolling a particular number on a die is 1/6 .<br> The total probability for all six outcomes equals one.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define the possible outcomes of a fair coin flip
outcomes = ['Heads','Tails']

# Calculate the PMF using the binomial distribution
probabilities = binom.pmf(np.arange(2), n=1, p=0.5)

# Plot the PMF
plt.figure(figsize=(4, 3))
plt.bar(outcomes, probabilities)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of a Fair Coin Flip')
plt.show()
```

### Probability Density Function : 
- **A Probability Density Function (pdf) is a mathematica function that describes a probability distribution of a continuous random variable.<br>**
- **A Random Variable which can take infinite number of values in an interval is known as Continuous Random Variable..<br>**
- **A pdf gives the probability density for different values of a continuous random variable..<br>**
- **It represents the relative likelihood that the variable takes on a particular value..<br>**
- **In other words, the pdf provides the likelihood of the random variable falling within a small range of values around
a specific point..<br>**
- **However, it doesnot give the probability of getting an exact value instead it gives the probabilty density at that point..<br>**
- **The area under the curve of a pdf over a certain interval represents the probability of the random variable falling within that interval..<br>**
- **The probability of a continuous random variable falling exactly at a single point is technically zero because the width of that point is infinitesimally small..<br>**

### Properties of PDF : 
**For a continuous random variable X, the pdf f(x) is a function that have the following properties :<br>**

**(1) The value of the pdf is always greater than or equal to zero for all real numbers.:<br>**
 \begin{equation*}f(x) \geq 0 \ , \quad \text{for all} \quad \ x \in \mathbb{R}\end{equation*}
**(2) f is piecewise continuous.<br>**

**(3) Total area under the pdf curve is always equal to one.:<br>**
 \begin{equation*}\int_{-\infty}^{\infty} f(x) \, dx = 1\end{equation*}
**(4) P(a≤X≤b):<br>**
 \begin{equation*}\int_{a}^{b} f(x) \, dx = 1\end{equation*}

**The first three conditions in the definition state the properties necessary for a function to be a valid pdf for a continuous random variable. The fourth condition tells us how to use a pdf to calculate probabilities for continuous random variables, which are given by integrals the continuous analog to sums.**

### Ans :-

### Cumulative Density Function (CDF) :
- **A CDF gives the probability that a random variable is less than or equal to a certain value.It is a cumulative function because it sums the total probabilities up to that point.It can be used to describe the probabilities for a discrete , continuous or mixed variable.<br>**
- **It provides the cumulative probability of observing a value less than or equal to a given point.<br>**
- **The cdf starts at zero and increases monotonically to 1 as the variable covers the entire range of possible values.<br>**
- **It gives the cumulative likelihood of observing a value up to a certain point , including all values below it.<br>**
- **The cdf is useful for determining probabilities associated with the range of values or finding percentiles of the distribution.<br>**
- **It is denoted as**

**Here X is the random variable and x is a specific value.The cdf gives us the probability that the random variable X is less than or equal to x. These functions are non-decreasing.As x increase , the probability can either increase or stay constant , but it cannot decrease**

### Why CDF is Used?
**The CDF helps us understand the cumulative probabilities associated with different values of a random variable.<br>**
**It’s useful for calculating percentiles, assessing risk, and making decisions based on probabilities.<br>**
**In statistical modeling, the CDF plays a crucial role in hypothesis testing, confidence intervals, and simulation studies.**

### Example :-
**For example, suppose we roll a dice one time. If we let x denote the number that the dice lands on, then the probability that the x is equal to different values can be described as follows:<br>**

**P(X=1): 1/6<br>**
**P(X=2): 1/6<br>**
**P(X=3): 1/6<br>**
**P(X=4): 1/6<br>**
**P(X=5): 1/6<br>**
**P(X=6): 1/6<br>**
**There is an equal chance that the dice could land on any number between 1 and 6.<br>**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

outcomes = np.arange(1, 7)

plt.figure(figsize=(4,3))

# Calculate the CDF 
cdf_values = randint.cdf(outcomes, 1, 7)

plt.plot(outcomes, cdf_values, marker='o', linestyle='-')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('CDF of a Fair Six-sided Die Roll')
plt.xticks(outcomes)
plt.grid(True)
plt.show()
```

### Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.
### Ans :-

- **Here, We're given a dataset that is normally distributed with a mean (μ) of 50 and a standard deviation (σ) of 10.<br>**

- **We need to find the probability that a randomly selected observation from this dataset will be greater than 60 <br>or Area under the curve ( > 60) .**

### Z-score : 
**First, we need to find , how many standard deviation 60 is away from the mean using Z-Score formula.<br>**

**Z-score = (x - μ) / σ = (60 - 50) / 10 = 1.0 <br>**
**Where:<br>**
**x is the value of your data point<br>**
**μ is the mean<br>**
**σ is the standard deviation<br>**

**--> Here, 60 is 1.0 standard deviation away from the mean.**

### Question : What is the probability that a random variable selected observation will be greater than 60 ?

- **Here , we have to find the Area Under the Curve ( > 60 )<br>**
- **Area Under the Curve means how much percentage of data falling over here ( > 60 )  in the form of percentage.<br>**
- **To find the value corresponding to 1.0 standard deviation in the Z-Score Table, we need to locate the value 1.0 in the table. In the table, the value corresponding to 1.0 is 0.84134, which represents the area less than 60.**

####  Area Under the Curve ( > 60 ) = 1 - 0.84134
####  &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp; &emsp;= 0.15866
#### So, the probability is approximately 0.1587 or 15.87%.
#### In other words, If we select any value from the given distribution then that value lie above 60 has a probablity of 15.87%. 

### Using Python Program

```python
from scipy.stats import norm

# Define the parameters
mean = 50
std_dev = 10
value = 60

# Calculate the z-score
z_score = (value - mean) / std_dev

# Calculate the probability using the cumulative distribution function (CDF)
probability = 1 - norm.cdf(z_score)

print("Probability:", probability)
```

### Q8: What is the z score? State the importance of the z score ?
### Ans :-

###  z-score :
- **In Statistics, A z-score measures how many standard deviations a data point is from the mean in a distribution.
It is also known as the standard score. Simply, it measures how far a data point is from the mean.<br>**

- **Z-score is measured in terms of standard deviations from the mean.<br>**

- **A z-score of 0 indicates that the data point’s score is the same as the mean score. A positive z-score indicates that the data point is above average, while a negative z-score indicates that the data point is below average.**

### Z Score Formula :
- **To calculate a z score, knowledge of the mean and standard deviation is required.<br>**

- **z is the z-score<br>**
- **x is the data point or value<br>**
- **μ is the mean value of the data set<br>**
- **σ is the standard deviation of the data set<br>**

### Example :

#### Suppose there are three students whose marks in their English examination <br>are 12, 16 and 23. The mean is 17.

```python
import math
marks = [12,16,23]
mean = sum(marks)/len(marks)
print('Mean:',mean)
# Here comes the standard deviation
# First let us calculate the individual deviations of observations from their mean
marks_dev = [abs((x-mean)**2) for x in marks]
st_dev = math.sqrt(sum(marks_dev)/len(marks_dev))
print('Standard deviation:',round(st_dev,2))
```

```python
# Calculating Z scores
Z_scores = [round((x-mean)/st_dev,2) for x in marks]
Z_scores
```

#### We’ve got the Z scores respectively for 12, 16 and 23 as -1.10, -0.22 and 1.32 respectively.

###  Importance of the z score :

### (3) Hypothesis testing:

The z-score can be used to calculate how far the sample mean deviates from the population mean, i.e. the normal distribution. This value can then be compared with a critical value of the normal distribution to determine whether the sample value differs significantly from the population value. If this is not the case, the hypothesis is accepted.<br>**

### (4) Confidence intervals: 
**The z-score is also used to calculate confidence intervals. This value of a sample mean can be used to estimate the range of values into which the population value falls with a certain probability.<br>**

### (5) Detection of outliers:
**Outlier is a value which differs significantly from other values in the data.<br>**

**By defining a threshold, outliers can be easily detected by calculating the z-score for each data point. Data whose values lie outside a defined range can be classified as outliers.<br>**

### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem ?
### Ans :-

###  Central Limit Theorem :
#### Definition :
**The Central Limit Theorem says that sampling distribution of the mean will always be normally distributed,<br> as long as the sample size is large enough.Regardless of whether the population has a normal, poisson, <br>binomial or any other distribution, the sampling distribution of the mean will be normal.<br>**

- **A sampling distribution for the sample mean is produced by repeatedly selecting simple random samples from the same population and of the same size, and then computing the sample mean for each of these samples. These samples are to be thought of as being independent of one another.<br>**

- **In CLT, Sampling should be random. The samples should not relate to one another.<br> One sample shouldn’t affect the others.<br>**

### Central Limit Theorem Formula :
**The shape of the sampling distribution of the mean can be determined without repeatedly sampling a population. The parameters are based on the population:**

### Conditions of the Central Limit Theorem :
**The central limit theorem states that the sampling distribution of the mean will always follow a normal distribution under the following conditions:<br>**

- **The sample size is sufficiently large. This condition is usually met if the size of the sample is n ≥ 30.<br>**
- **The samples are independent and identically distributed, i.e., random variables. The sampling should be random..<br>**
- **The population’s distribution has a finite variance. The central limit theorem doesn’t apply to distributions with infinite variance..<br>**

### Why n ≥ 30 Samples?
**The sample size of 30 is considered sufficient to see the effect of the CLT. If the population distribution is closer to the normal distribution, you will need fewer samples to demonstrate the central limit theorem. On the other hand, if the population distribution is highly skewed, you will need a large number of samples to understand the CLT.**

### Significance of the Central Limit Theorem :
**The central limit theorem has both, statistical significance as well as practical applications.**

### (1) Statistical Significance of CLT :

- **Analyzing data involves statistical methods like `hypothesis testing` and constructing `confidence intervals`. These methods assume that the population is normally distributed. In the case of unknown or non-normal distributions, we treat the sampling distribution as normal according to the central limit theorem.<br>**

- **If we increase the samples drawn from the population, the standard deviation of sample means will decrease. This helps us estimate the mean of the population much more accurately. Also, the sample mean can be used to create the range of values known as a confidence interval (that is likely to consist of the population mean).**

### (2) Practical Applications of CLT :
**The central limit theorem has many applications in different fields.**

- **`Political/election polls` are prime CLT applications. These polls estimate the percentage of people who support a particular candidate. You might have seen these results on news channels that come with confidence intervals. The central limit theorem helps calculate the same.<br>**

- **Confidence interval, an application of CLT, is used to calculate the mean family income for a particular region.**

- **Samples should be independent of each other. One sample should not influence the other samples.<br>**

### (ii) Sample Size :
- **Sample size should be not more than 10% of the population when sampling is done without replacement.<br>**
- **The sample size should be sufficiently large.<br>**
- **The sample size of 30 is considered sufficient to see the effect of the CLT. If the population distribution is closer to the normal distribution, you will need fewer samples to demonstrate the central limit theorem. On the other hand, if the population distribution is highly skewed, you will need a large number of samples (means  `n ≥ 30`)  to understand the CLT.<br>**

### (iii) Finite Variance : 
- **The population from which the sample is drawn should have a finite variance (i.e., the variance should not be infinite).<br>**
- **Infinite variance can lead to issues with convergence.**

### Important Note :
**The central limit theorem has important implications in applied machine learning. This theorem does inform the solution to linear algorithms such as linear regression, but not for complex models like artificial neural networks(deep learning) because they are solved using numerical optimization methods.**
